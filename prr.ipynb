```python
import matplotlib.pyplot as plt

with open("prr_results.txt", "r") as file:
    prr_values = [float(line.strip()) for line in file]

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(prr_values) + 1), prr_values, marker='o', linestyle='-', color='b')
plt.title("Packet Reception Ratio (PRR) per Simulation Run")
plt.xlabel("Run Number")
plt.ylabel("Packet Reception Ratio (PRR) %")
plt.grid(True)

average_prr = sum(prr_values) / len(prr_values)
plt.axhline(y=average_prr, color='r', linestyle='--', label=f'Average PRR: {average_prr:.2f}%')
plt.legend()

plt.show()
```
